```python
from pydub import AudioSegment
import os

# ffmpeg மற்றும் ffprobe-இன் முழுப்பாதைகளை அமைத்தல்
AudioSegment.converter = r"C:\Users\ajisha\Downloads\ffmpeg-7.1-win64\bin\ffmpeg.exe"
AudioSegment.ffprobe = r"C:\Users\ajisha\Downloads\ffmpeg-7.1-win64\bin\ffprobe.exe"

# MP3 கோப்பு இருப்பதைச் சரிபார்த்தல்
if os.path.exists("new.mp3"):
    print("கோப்பு கிடைக்கிறது")
    sound = AudioSegment.from_mp3("new.mp3")
    from pydub.playback import play
    play(sound)
else:
    print("கோப்பு கிடைக்கவில்லை")
```
